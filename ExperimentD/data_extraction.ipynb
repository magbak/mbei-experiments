```python
import pandas as pd
import jsonlines
pd.set_option('display.max_rows',1000)
```

```python
files = ["stderr_2022_03_10_13_00_00_13_59_59_S0.json", "stderr_2022_03_10_13_00_00_13_59_59_S1.json"]
```

```python
loglines = []
for f in files:
    with jsonlines.open(f) as reader:
        for obj in reader:
            if "textPayload" in obj:
                loglines.append(obj["textPayload"])
```

```python
df = pd.DataFrame({"logs":loglines})
```

```python
df = df[df["logs"].str.contains(" message processing took ")].copy()
```

```python
df["microsecs"] = df["logs"].str.extract("took (.*) μs")
```

```python
df["microsecs"] = df["microsecs"].astype(int)
```

```python
df["microsecs"].all()
```

```python
df["first_update"] = df["logs"].str.extract("μs, first update: (.*), n_deltas")
```

```python
df["n_deltas"] = df["logs"].str.extract("n_deltas: (.*), n_events").astype(int)
df["n_events"] = df["logs"].str.extract("n_events: (.*), n_retractions").astype(int)
df["n_retractions"] = df["logs"].str.extract("n_retractions: (.*), n_reprocessing").astype(int)
df["n_reprocessing"] = df["logs"].str.extract("n_reprocessing: (.*), n_open_edges").astype(int)
df["n_open_edges"] = df["logs"].str.extract("n_open_edges: (.*)").astype(int)
```

```python
df["n_retractions"].sum()
```

```python
df[df["first_update"] == "deltas"].sort_values(by="microsecs", ascending=False).head(1000)
```

```python
df[df["first_update"] == "event"].sort_values(by="n_reprocessing", ascending=False)
```

```python
df[df["first_update"] == "retractions"].sort_values(by="n_reprocessing", ascending=False)
```
